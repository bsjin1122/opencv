## 에지 검출 이해를 돕습니다. 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/sudoku.jpg', 0)
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F, 1, 0, ksize=5) #64F 플로팅 포인트 정확성  # x방향의 도함수, y방향의 도함수
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F, 0, 1, ksize=5) #64F 플로팅 포인트 정확성  # dx x방향의 도함수, dy y방향의 도함수
```

```python
display_img(sobely)
```

## 라플라시안 도함수

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
blended = cv2.addWeighted(src1 = sobelx, alpha=0.5, src2=sobely, beta=0.5, gamma=0)
```

```python
display_img(blended)
```

```python
ret, th1 = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
```

```python
display_img(gradient)
```
