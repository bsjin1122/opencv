## 이미지를 HSV 또는 HSL로 변환하는 방법

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg')
plt.imshow(img)
```

```python
img =cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
